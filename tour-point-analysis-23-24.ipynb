```python
import pandas as pd
import matplotlib.pyplot as plt

df_tournaments = pd.read_csv('tour_events.csv')
df_tournaments.head()
# clean all cells by removing spacing
df_tournaments.columns = df_tournaments.columns.str.strip()
```

```python
# Chart North island events vs South island events
df_tournaments['Island'].value_counts().plot(kind='bar')
```

```python
# Chart Tiers 
df_tournaments['Tier'].value_counts().plot(kind='bar')
```

```python
df_tournaments['Points_allocated'].value_counts().plot(kind='bar')
```

```python
df_tournaments['Rounds'].value_counts().plot(kind='bar')
```

```python
df_tournaments['Holes per round'].value_counts().plot(kind='bar')
```

```python
# Chart players per event in categories of 10, sort by low to high
df_tournaments['Players'].value_counts(bins=10).sort_index().plot(kind='bar')
```
